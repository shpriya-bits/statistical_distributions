### **BI-VARIATE ANALYSIS**

```
sns.relplot(x='sugarpercent',y='pricepercent',data=candy_data)
```

```
sns.relplot(x='winpercent',y='pricepercent',data=candy_data, hue='winpercent')
```

```
sns.catplot(x='chocolate',y='pricepercent',data=candy_data, kind="swarm" )
```

```
#analysing categorical data against the price percent

sns.catplot(x='fruity',y='pricepercent',data=candy_data, kind="swarm" )
sns.catplot(x='chocolate',y='pricepercent',data=candy_data, kind="swarm" )
```

### **MULTI-VARIATE ANALYSIS**

```
sns.relplot(x='sugarpercent',y='winpercent',data=candy_data, size='pricepercent', hue='pricepercent')
```

```
sns.lmplot(x='pricepercent',y='winpercent', hue="chocolate", data=candy_data )
```

```
#from the above multi variate analysis we could see a correlation between the wine percent and the price percent.
#this will be used to derive a model in predicting the candy price based on the winpercent
#this can be used to predict the price of the candy.
```
